# EXPLORATORY DATA ANALYSIS

```python
df['discounted_price'] = df['discounted_price'].replace('[\₹,]', '', regex=True)
df['discounted_price'] = df['discounted_price'].astype('float')
```

```python
salecat=df.groupby(['category'],as_index=False)['discounted_price'].sum().sort_values(by='discounted_price',ascending=False)
plt.figure(figsize=(10, 6)) 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
sns.barplot(x='category',y='discounted_price',data=salecat.head(10))
```

#### We can observe that the most sales are in television and smart phones

```python
df['actual_price'] = df['actual_price'].replace('[\₹,]', '', regex=True)
df['actual_price'] = df['actual_price'].astype('float')
df['actual_price'].plot(kind='hist')
```

```python
df['rating_count'] = df['rating_count'].replace(',', '', regex=True).astype(int)
df['rating_count'] = df['rating_count'].astype('int')
top_10_products = df.sort_values(by='rating_count', ascending=False).head(10)
plt.figure(figsize=(10, 6)) 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
sns.barplot(x='product_name',y='rating_count', data=top_10_products)
```

##### From the above data we can infer that amazon basics ethernet cables sales are high considering reviews customers say its cost effective and has a good quality so the company may prefer manufacturing more such cost effctive products in the area of cables.

```python
salecats=df.groupby(['category'],as_index=False)['rating_count'].sum().sort_values(by='rating_count',ascending=False)
plt.figure(figsize=(10, 6)) 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
sns.barplot(x='category',y='rating_count',data=salecats.head(10))
```
